# Intro sur seaborn

C'est une surcouche à Matplotlib adaptée aux DataFrame et qui fait tout, même la vaisselle. 

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(style="whitegrid", palette="muted")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue="species")
```

```python
iris2 = pd.melt(iris, "species", var_name="measurement")
sns.swarmplot(x="measurement", y="value", hue="species", data=iris2, size=2)
```

```python
# Load the example tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True,
               inner="quart", palette={"Male": "b", "Female": "y"})
sns.despine(left=True)
```
